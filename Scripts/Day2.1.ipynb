```python
# function to load document
def load_doc(filename):
    #open the file as read only
    file = open(filename, 'rt')
    # read all text
    text = file.read()
    # close the file
    file.close()
    return text
```

```python
# turn a doc into clean tokens
def clean_doc(doc):
    # split text into tokens
    tokens = doc.split()
    # convert to lower case
    #words_lo = [word.lower() for word in tokens]
    # remove punctuation from each word
    re_punc = re.compile('[%s]' % re.escape(string.punctuation))
    tokens = [re_punc.sub('', w) for w in tokens]
    # remove all tokens that are not alphabetic
    tokens = [word for word in tokens if word.isalpha()]
    # filter out stop words
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if not word in stop_words]
    # filter out short tokens
    tokens = [word for word in tokens if len(word) > 1]
    return tokens
```

```python
# load doc and add to vocab
def add_doc_to_vocab(filename, vocab):
    # load doc
    doc = load_doc(filename)
    # clean doc
    tokens = clean_doc(doc)
    # update counts
    vocab.update(tokens)
```

```python
from collections import Counter
#load all docs in a directory
def process_docs2(directory, vocab):
    
    # walk through all files in the folder
    for filename in os.listdir(directory):
        #skip files that do not have the right extension
        if not filename.endswith('.txt'):
            next
        # create the full path of the file to open
        path = directory + '/' + filename
        #add doc to vocab
        add_doc_to_vocab(path, vocab)

```

```python
# define vocab
vocab = Counter()
# add all docs to vocab
process_docs2('txt_sentoken/neg', vocab)
process_docs2('txt_sentoken/pos', vocab)
# print the size of the vocab
print(len(vocab))
```

```python
# keep tokens with > 5 occurences
def occurence(vocab, occur=5):
    tokens = [k for k,c in vocab.items() if c >= occur]
    return tokens
```

```python
# define functions to 
def save_list(lines, filename):
    data = '\n'.join(lines)
    file = open(filename, 'w')
    file.write(data)
    file.close()
```

```python
# Define vocab
#directory = 'NLP-studying'
vocab = Counter()
process_docs2('txt_sentoken/neg', vocab)
process_docs2('txt_sentoken/pos', vocab)
token = occurence(vocab, 6)
path = 'vocab_' + str(len(token)) + '.txt'
save_list(token, path)
print(len(token))
```

```python
# load doc, clean and return line of tokens
def doc_to_line(filename, vocab):
    # load the doc
    doc = load_doc(filename)
    # clean doc
    tokens = clean_doc(doc)
    # filter by vocab
    tokens = [w for w in tokens if w in vocab]
    return ' '.join(tokens)
```

```python
# load all docs in a directory
def process_docs3(directory, vocab):
    lines = list()
    
    # walk through all files in the folder
    for filename in os.listdir(directory):
        #skip files that do not have the right extension
        if not filename.endswith('.txt'):
            next
        # create the full path of the file to open
        path = directory + '/' + filename
        # load and clean the doc
        line = doc_to_line(path, vocab)
        #add to list
        lines.append(line)
    return lines
```

```python
# load vocabulary
vocab_filename = 'vocab_27139.txt'
vocab = load_doc(vocab_filename)
vocab = vocab.split()
vocab = set(vocab)
len(vocab)
```

```python
# prepare negative reviews
negative_lines = process_docs3('txt_sentoken/neg', vocab)
save_list(negative_lines, 'negative.txt')
print(len(negative_lines))
# prepare positive reviews
positive_lines = process_docs3('txt_sentoken/pos', vocab)
save_list(positive_lines, 'positive.txt')
print(len(positive_lines))
```
