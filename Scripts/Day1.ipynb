### Multilayer Perceptron

```python
# import modules
from keras.utils import plot_model
from keras.models import Model
from keras.layers import Input
from keras.layers import Dense
visible = Input(shape = (10, ))
hidden1 = Dense(10, activation='relu')(visible)
hidden2 = Dense(20, activation='relu')(hidden1)
hidden3 = Dense(10, activation='relu')(hidden2)
output_layer = Dense(1, activation='sigmoid')(hidden3)
model = Model(inputs=visible, outputs=output_layer)
# summarize layers
model.summary()
# plot graph
plot_model(model)
```

### Covolutional Neural Network

```python
# Import modules
from keras.layers.convolutional import Conv2D
from keras.layers.pooling import MaxPooling2D
visible_layer = Input(shape=(64, 64, 1))
conv1 = Conv2D(32, kernel_size=4, activation='relu')(visible_layer)
pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)
conv2 = Conv2D(16, kernel_size=4, activation='relu')(pool1)
pool2 = MaxPooling2D(pool_size=(2, 2))(conv2)
full_layer = Dense(10, activation='relu')(pool2) 
outputlayer = Dense(1, activation='sigmoid')(full_layer)
model2 = Model(inputs=visible_layer, outputs=outputlayer)
# summarize layers
model2.summary()
#plot graph
plot_model(model2)
```

### Recurrent Neural Network


```python
# import modules
from keras.layers.recurrent import LSTM
visible3 = Input(shape=(100, 1))
hidden_1 = LSTM(10)(visible3)
hidden_2 = Dense(10, activation='relu')(hidden_1)
output3 = Dense(1, activation='sigmoid')(hidden_2)
model3 =  Model(inputs=visible3,outputs=output3)
# summarize layers
model3.summary()
# plot graph
plot_model(model3)
```

### Functional API

```python
# Import modules
from keras.utils import plot_model
from keras.models import Model
from keras.layers import Input, Dense, Flatten
from keras.layers.convolutional import Conv2D
from keras.layers.pooling import MaxPooling2D
from keras.layers.merge import concatenate
# Input layer
visible_ = Input(shape=(64,64,1))
# First feature extractor
conv2 = Conv2D(32, kernel_size=4, activation='relu')(visible_)
pool1 = MaxPooling2D(pool_size=(2, 2))(conv2)
flat1 = Flatten()(pool1)
# Second feature extractor
conv2 = Conv2D(16, kernel_size=8, activation='relu')(visible_)
pool2 = MaxPooling2D(pool_size=(2,2))(conv2)
flat2 = Flatten()(pool2)
# Merge feature extractors
merge = concatenate([flat1, flat2])
# Interpretation layer
hidden1 = Dense(10, activation='relu')(merge)
# Prediction output
output4 = Dense(1, activation='sigmoid')(hidden1)
model4 = Model(inputs=visible_, outputs=output4)
# Summarize layers
print(model4.summary())
# Plot graph
plot_model(model4)
```
